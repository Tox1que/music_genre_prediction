Задача – разработать модель, позволяющую классифицировать музыкальные произведения по жанрам.

```python
title_dict = {
    'instance_id': [0, 'идентификатора'],
    'music_genre': ['Музыкальный жанр', 'жанров'],
    'key': ['Базовый ключ (нота) произведения', 'нот'],
    'mode': ['Модальность (мажорная или минорная) трека', 'модальности'],
    'obtained_date': ['Дата загрузки в сервис', 'дат'],
    'acousticness': ['Акустичность', 'акустичности'],
    'danceability': ['Танцевальность', 'танцевальности'],
    'duration_ms': ['Продолжительность', 'продолжительности (в милисекундах)'],
    'energy': ['Энергичность', 'энергичности'],
    'instrumentalness': ['Инструментальность', 'инструментальности'],
    'liveness': ['Привлекательность', 'привлекательности'],
    'loudness': ['Громкость', 'громкости'],
    'speechiness': ['Выразительности', 'выразительности'],
    'tempo': ['Темп', 'темпа'],
    'valence': ['Привлекательность для пользователей',
                'привлекательности для пользователей']
}


def get_overview_histogram(df, col, ax, bins_num='auto', line_color='r',
                           color=None):
    ax.set_title(f'Распределение {title_dict[col][1]}')
    median = df[col].median()
    mean = df[col].mean()
    sns.histplot(df[col], kde=True, color=color, bins=bins_num, ax=ax)
    ax.axvline(x=median, c=line_color, label='Медиана')
    ax.axvline(x=mean, c=line_color, ls='--', label='Среднее')
    ax.set_ylabel('Количество')
    


def get_overview(df_1, df_2, cols):
    fig, axs = plt.subplots(3, 4, figsize=(15, 7))
    fig.delaxes(axs[2][3])
    for col, ax in tqdm(zip(cols, axs.flat)):
        get_overview_histogram(df_1, col, ax)
        get_overview_histogram(df_2, col, ax, line_color='black',
                               color='sandybrown')
    plt.tight_layout()
    plt.legend(['kde обучающей выборки', 'медиана обучающей выборки',
                'среднее обучающей выборки', 'kde тестовой выборки',
                'медиана тестовой выборки', 'среднее тествовой выборки', 
                'Распределение обучающей выборки',
                'Распределение тестовой выборки'],
               bbox_to_anchor=(2.3, 1), fontsize=11)
    plt.show()


def get_histogram(df, col, bins_num='auto', fig_size=(15, 5)):
    fig, axs = plt.subplots(2, 1, figsize=fig_size, sharex=True,
                            gridspec_kw={'height_ratios': [3, 1]})
    get_overview_histogram(df, col, axs[0], bins_num)
    red_circle = dict(
        markerfacecolor='red', marker='o', markeredgecolor='white'
    )
    axs[0].legend()
    ax = axs[1]
    df[col].plot.box(flierprops=red_circle, ax=ax, vert=False)
    ax.set_yticks([])
    ax.set_xlabel(title_dict[col][0])
    plt.tight_layout()
```

## Исследовательский анализ

```python
def get_missing_values_percentage(df):
    percent_missing = df.isna().mean() * 100
    percent_missing.sort_values(inplace=True, ascending=False)
    return percent_missing.drop(percent_missing[percent_missing == 0].index)


def get_missing_values_barplot(df):
    temp = get_missing_values_percentage(df)
    plt.figure(figsize=(10, 2))
    pal = sns.color_palette('Blues', len(temp))
    ax = sns.barplot(y=temp.index, x=temp.values,
                     palette=pal[::-1], dodge=False)
    plt.ylabel('')
    text = 'обучающая выборка'
    if df.shape[0] < 1e4:
        text = 'тестовая выборка'
    plt.title(f'Процент пропущенных значений по стобцам ({text})')
    plt.yticks(fontsize=12)
    for item in ax.patches:
        if not np.isnan(item.get_width()):
            plt.text(item.get_width(), item.get_y(),
                     f'{item.get_width():.1f}%', va='top',
                     fontsize=12)
```

```python
def get_temp(df, col):
    temp = df.pivot_table(index='music_genre', values=col, aggfunc=['median'])
    temp.columns = ['median']
    temp = temp.sort_values(by='median')
    return temp


def get_boxplot(df, col, f_dec, fig_size=(15, 5)):
    temp = get_temp(df, col)
    plt.figure(figsize=fig_size)
    order = temp.index
    circles = dict(markerfacecolor='cadetblue',
                   marker='o', markeredgecolor='white')
    sns.boxplot(data=df, x='music_genre', y=col, showmeans=True, order=order,
                palette='ch:s=.25,rot=-.25', medianprops=dict(color='w'),
                flierprops=circles, linewidth=0.5)
    for i in range(temp.shape[0]):
        genre_median = temp['median'][i]
        plt.text(i-0.15, genre_median+genre_median*0.01,
                 f'{genre_median:{f_dec}f}', color='w', fontweight='bold')
    plt.ylabel(title_dict[col][0])
    plt.xlabel('Музыкальный жанр')
    plt.title(f'Значения {title_dict[col][1]} по жанрам')


def get_barplot(df, col, fig_size=(15, 3)):
    temp = df[col].value_counts().to_frame()
    temp.columns = ['count']
    plt.figure(figsize=fig_size)
    pal = sns.color_palette('ch:s=.25,rot=-.25', temp.shape[0])
    sns.barplot(data=temp, x=temp.index, y='count', hue='count', alpha=.5,
                dodge=False, palette=pal)
    percentage = df[col].value_counts(normalize=True) * 100
    index_value_percentage = zip(range(temp.shape[0]), temp.values,
                                 percentage.values)
    for idx, value, percentage in index_value_percentage:
        plt.text(idx, value, f'{percentage:.2f}%')
    plt.ylabel('Количество')
    plt.gca().legend_.remove()
    plt.xlabel('Музыкальный жанр')
    plt.title('Количество треков по жанрам')
```

```python
def get_cat_temp(df, col, target):
    temp = df.pivot_table(index=[col, target], values='track_name',
                          aggfunc='count')
    temp = temp.unstack(level=1)
    return temp


def get_cat_barplot(df, target, col='music_genre', fig_size=(15, 5), rot=0,
                    c_map='ocean'):
    local_dict = {'key': 'Тональность', 'mode': 'Модальность'}
    temp = get_cat_temp(df, col, target)
    temp.plot(kind='bar', cmap=c_map, figsize=(15, 5),
              alpha=.65, stacked=True)
    handles, labels = plt.gca().get_legend_handles_labels()
    labels = [label.split(', ')[-1].rstrip(')') for label in labels]
    if target == 'mode':
        temp.columns = ['Major', 'Minor']
        percentage = temp['Major'] / (temp['Major'] + temp['Minor']) * 100
        genre_mode_perc = zip(range(temp.shape[0]), temp['Major'], percentage)
        for genre, major, percentage in genre_mode_perc:
            plt.text(genre, major, f'{percentage:.0f}%', ha='center')

    plt.legend(handles[::-1], labels[::-1], bbox_to_anchor=(1, 1),
               title=local_dict[target])
    plt.ylabel('Количество')
    plt.xlabel(title_dict[target][0])
    plt.xticks(rotation=rot)
    plt.title(f'Распределение {title_dict[target][1]} по жанрам')
```

```python
def clean_title(df):
    '''
    Убирает из названия трека знаки пунктуации.
    '''
    df['name_clean'] = (
        df['track_name'].str.lower()
        .replace(f'[{punctuation}]', '', regex=True)
    )
    df['name_clean'] = df['name_clean'].apply(lambda x: ' ' + x + ' ')

    return df['name_clean']
```

```python
genre_column = {
    'Anime': 'anime_1', 'Electronic': 'electronic_1', 'Classical': 'classic_1',
    'Rock': 'rock_1', 'Rap': 'rap_1', 'Alternative': 'alternative',
    'Country': 'country', 'Blues': 'blues_1', 'Jazz': 'jazz', 'Hip-Hop': 'hip'
}


def get_most_common_words_by_genre(df, top_words_num=50):
    '''
    Для каждого жанра выбирает топ 50 слов,
    наиболее часто встречаемых в названии трека.
    '''
    genre_words = {}
    for genre in df['music_genre'].unique():
        result = Counter()
        (
            df[df['music_genre'] == genre]['name_clean']
            .str.lower().str.split().apply(result.update)
        )
        result = (
            pd.DataFrame(result.items())
            .sort_values(by=1, ascending=False)[0].head(top_words_num).values
        )
        genre_words[genre] = result
    return genre_words


def get_unique_words(df, top_words_num):
    '''
    Из топ 50-ти слов, выбранных для каждого жанра,
    оставляет только уникальные.
    '''
    genre_words = get_most_common_words_by_genre(df, top_words_num)
    unique_words = set([x for item in genre_words.values() for x in item])
    return unique_words


def get_expression(df, genre, top_words_num=50):
    '''
    Возвращает регулярное выражение, содержащее слова,
    которые наиболее часто будут встречаться именно в заданном жанре.
    '''
    unique_words = get_unique_words(df, top_words_num)
    expression = []
    for word in tqdm(unique_words):
        value_counts = (
            df[df['name_clean']
            .str.contains(f'[^A-z]{word}[^A-z]', case=False)]
            ['music_genre'].value_counts()
        ).index.to_list()
        if value_counts[0] == genre:
            expression.append(word)
    expression = '[^A-z]|[^A-z]'.join(expression)
    expression = '[^A-z]' + expression + '[^A-z]'
    return expression
```

```python
def add_column_to_df(df, df_temp, col, expr, main_col):
    '''
    Добавляет новый признак в основной набор данных 
    и удаляет обработанные объекты из временного датасета.
    '''
    df_temp[col] = df_temp[main_col].str.contains(expr, case=False)
    df[col] = False
    indices = df_temp.query(f'{col} == True').index
    df.loc[indices, col] = True
    df[col] = df[col].astype('int8')
    df_temp = df_temp.drop(indices, axis=0)
    return df, df_temp


def add_column(df, df_test, df_temp, df_test_temp,
               col, expr, main_col='track_name'):
    df, df_temp = add_column_to_df(df, df_temp, col, expr, main_col)
    df_test, df_test_temp = add_column_to_df(df_test, df_test_temp,
                                             col, expr, main_col)
    return df, df_test, df_temp, df_test_temp
```

```python
def get_target_features_na_df(df, col):
    notna_df = df.loc[df[col].notna()]
    na_df = df.loc[df[col].isna()]
    if col == 'duration_ms':
        notna_df = df.loc[df[col] > -1]
        na_df = df.loc[df[col] == -1]
    features = notna_df.drop(col, axis=1)
    target = notna_df[col]
    return target, features, na_df
```

```python
def objective(trial, X, y, ohe_cols, encoder=oh_encoder, weight=None,
              classifier='rf', metric='f1_micro', l_f=None,
              depth=(2, 30), split=(2, 35), features=(2, 32),
              c=(7.5, 8.7), slvr=['lbfgs'], mx_itr=(100, 3000),
              intrcp_sclng=(45, 50), tl=(.25, .35)):
    transformer = make_column_transformer((encoder, ohe_cols),
                                           remainder='passthrough')
    if classifier == 'rf':
        rf_params = {
            'n_estimators': trial.suggest_int('n_estimators', 20, 100),
            'max_depth': trial.suggest_int('max_depth', *depth),
            'max_features': trial.suggest_int('max_features', *features),
            'min_samples_split': trial.suggest_int('min_samples_split',
                                                   *split)
        }
        clf = RandomForestClassifier(n_jobs=-1, random_state=SEED,
                                     class_weight=weight, **rf_params)

    elif classifier == 'cat':
        param = {
            'colsample_bylevel': trial.suggest_float(
                'colsample_bylevel', 0.01, 1
            ),
            'depth': trial.suggest_int('depth', 2, 5),
            'boosting_type': trial.suggest_categorical(
                'boosting_type', ['Ordered', 'Plain']
            ),
            'bootstrap_type': trial.suggest_categorical(
                'bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS']
            ),
            'used_ram_limit': '3gb'
        }
        clf = CatBoostClassifier(**param, auto_class_weights=weight,
                                 silent=True, random_state=SEED,
                                 loss_function=l_f, thread_count=-1)

    else:
        lr_params = {
            'C': trial.suggest_float('C', *c),
            'max_iter': trial.suggest_int('max_iter', *mx_itr),
            'solver': trial.suggest_categorical(
                'solver', slvr            
            ),
            'intercept_scaling': trial.suggest_int(
                'intercept_scaling', *intrcp_sclng
            ),
            'tol': trial.suggest_float('tol', *tl)
        }
        clf = LogisticRegression(**lr_params, class_weight=weight,
                                 random_state=SEED, n_jobs=-1)

    clf = make_pipeline(transformer, StandardScaler(), clf) 
    return cross_val_score(
        clf, X, y, scoring=metric, n_jobs=-1,
        cv=StratifiedKFold(5, random_state=SEED, shuffle=True)
    ).mean()
```

```python
def get_predictions(col, ohe_cols, features, target, na_df,
                    params={}, encoder=oh_encoder,
                    weight='balanced'):
    transformer = make_column_transformer(
        (oh_encoder, ohe_cols), remainder='passthrough'
    )
    model = RandomForestClassifier(**params, random_state=SEED,
                                   class_weight=weight)
    if col in ('tempo', 'duration_ms'):
        model = LinearRegression(normalize=True)
    model = make_pipeline(transformer, StandardScaler(),  model)
    model.fit(features, target)
    
    na_features = na_df.drop(col, axis=1)
    predicted = model.predict(na_features)
    return predicted
```

Лучший результат у модели градиентного бустинга – воспользуемся ею в качестве мета-модели в стекинге. 

```python
def get_stacking_clf(bst_prms, lr_bst_prms, l_f='MultiClass'):
    level_0 = []
    level_0.append(('lr', LogisticRegression(**lr_bst_prms, n_jobs=-1,
                                             random_state=SEED)))
    level_0.append(
        (('rf'), RandomForestClassifier(random_state=SEED, n_jobs=-1))
    )
    level_0.append(
        (('cat'), CatBoostClassifier(
            random_state=SEED, loss_function=l_f,
            silent=True, thread_count=-1
        ))
    )
    level_1 = CatBoostClassifier(
        **bst_prms, silent=True, loss_function=l_f,
        random_state=SEED, thread_count=-1
    )
    clf = StackingClassifier(estimators=level_0, final_estimator=level_1,
                             n_jobs=-1)
    return clf
```

```python
def objective(trial, X, y, encoder=oh_encoder,
              lr_bst_prms=lr_best_params,
              cols=ohe_columns, metric='f1_micro', l_f='MultiClass'):

    transformer = make_column_transformer((encoder, cols),
                                           remainder='passthrough')

    params = {
        'colsample_bylevel': trial.suggest_float(
            'colsample_bylevel', 0.01, 1
        ),
        'depth': trial.suggest_int('depth', 2, 5),
        'boosting_type': trial.suggest_categorical(
            'boosting_type', ['Ordered', 'Plain']
        ),
        'bootstrap_type': trial.suggest_categorical(
            'bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS']
        ),
        'used_ram_limit': '3gb'
    }

    if params['bootstrap_type'] == 'Bayesian':
        params['bagging_temperature'] = trial.suggest_float(
            'bagging_temperature', 0, 10
        )
    elif params['bootstrap_type'] == 'Bernoulli':
        params['subsample'] = trial.suggest_float('subsample', 0.1, 1)

    clf = get_stacking_clf(params, lr_bst_prms)

    clf = make_pipeline(transformer, StandardScaler(), clf)
    return cross_val_score(
        clf, X, y, scoring=metric, n_jobs=-1,
        cv=StratifiedKFold(5, random_state=SEED, shuffle=True)
    ).mean()
```

```python
def plot_permutation_importance(clf, cols=columns, fwr=features_with_random, 
                           trgt=target, nrepeats=10):
    first_char = str(clf)[0]
    if first_char == '<':
        fwr = fwr.values
    result = permutation_importance(clf, fwr, trgt, scoring='f1_micro',
                                    n_repeats=nrepeats, random_state=SEED)
    importances = pd.Series(data=result.importances_mean,
                            index=cols).sort_values()
    plt.figure(figsize=(15, 10))
    plt.barh(importances.index, importances.values)
    local_dict = {'L': 'модели логистической регрессии',
                  'R': 'модели случайного леса (гиперпараметры по умолчанию)',
                  '<': 'мета-модели'}
    clf_name = local_dict[first_char]
    plt.title(f'Оценка важности признаков для {clf_name} путём перестановок')
```
